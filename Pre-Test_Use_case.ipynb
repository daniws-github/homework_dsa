### Q3

```python
total = df.shape[0]
churn = df.groupby('class').number_customer_service_calls.count()[1]

churn_persen = churn / total

print('{:.2%} of the base are classified as churn'.format(churn_persen))
```

### Q4

```python
cs_call3x = df[df['number_customer_service_calls'] >= 3]
x,y = cs_call3x.groupby('class').state.count()

z = y / (x+y)

print('{:.2%} of the customers who called the company more than 3 times are classified as Churn'.format(z))
```

### Q5

```python
ir_call = df[df['international_plan'] == 0]
x,y = ir_call.groupby('class').state.count()
z = y / (x+y)

print('{:.2%} of those with no international plan are classified as Churn'.format(z))
```
